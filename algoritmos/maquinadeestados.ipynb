```python


estado = 0
botao = 0
sensor_porta = 0

while True:
  botao = int(input("Botao: "))
  sensor_porta = int(input("Sensor da Porta: "))
  '''LOGICA DE TRANSIÇÃO'''
  if estado == 0:
    if botao == 1:
      estado = 1
  elif estado == 1:
    if sensor_porta == 1:
      estado = 2
  elif estado == 2:
    if botao == 1:
      estado = 0

  '''LOGICA DE AÇÃO'''
  if estado == 0:
    buzina = 0
    luz_vermelha = 0
  if estado == 1:
    buzina = 0
    luz_vermelha = 1
  if estado == 2:
    buzina = 1
    luz_vermelha = 1
  if estado == 3:
    buzina = 0
    luz_vermelha = 0
  print(f'{buzina}')
  print(f'{luz_vermelha}')
```
